# Replace -> fillna()

```python
df['SpeciesIDDesc'].value_counts()
```

```python
df['SpeciesIDDesc'].fillna(value="OTHER", inplace=True)
```

```python
df['SpeciesIDDesc'] = df['SpeciesIDDesc'].fillna(value="OTHER")
```
